```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.feature_selection
from sklearn.feature_selection import mutual_info_classif

file = 'heart.csv'
read = pd.read_csv( file )
data = pd.DataFrame( read )
data
```

# Age

```python

data_x = data.drop( columns=['output'] )
mutual_info = mutual_info_classif( data_x, data['output'], discrete_features='auto' )
mi_scores = pd.Series( mutual_info, index=data_x.columns )
mi_scores
```
